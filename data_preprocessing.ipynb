```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

Modificam randurile cu valori lipsa in coloanele categoriale, inlocuindu-le cu valoarea cea mai frecventa din fiecare coloana

```python
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num',
                'marital-status', 'occupation', 'relationship', 'race', 'sex',
                'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']

df = pd.read_csv('adult.data', names=column_names, na_values='?', skipinitialspace=True)
for col in ['workclass', 'occupation', 'native-country']:
    df[col] = df[col].fillna(df[col].mode()[0])
```

Aplicam transformare logaritmica pentru a repara asimetria pe capital-gain si capital-loss

```python
df['capital-gain'] = np.log1p(df['capital-gain'])
df['capital-loss'] = np.log1p(df['capital-loss'])
```

Modificam feature-urile ca native-country, income si sex ca valori binare

```python
df['native-country'] = df['native-country'].apply(lambda x: 1 if x == 'United-States' else 0)
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
df['sex'] = df['sex'].map({'Male': 1, 'Female': 0})
```

Stergem fnlwgt deoarece nu este relevant pentru predictie si education deoarece este corelata puternic cu education-num

```python
df = df.drop(['fnlwgt', 'education'], axis=1)
```

Folosim One-Hot Encoding pentru a transforma variabilele categoriale

```python
df = pd.get_dummies(df, drop_first=True)
print(df.head())
```

Folosim StandardScaling pentru a scala variabilele numerice

```python
scaler = StandardScaler()
num_cols = ['age', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df.to_csv('adult_preprocessed.csv', index=False)
```
